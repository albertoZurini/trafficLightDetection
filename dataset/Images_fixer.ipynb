```python
import numpy as np
import cv2
from os import listdir, mkdir
from os.path import isfile, isdir, join
```

```python
width, height = 72, 98 # px
imgsDir = 'imgs/' # path of raw images
dstDir = 'imgs_fixed/' # path of resized images

if not isdir(dstDir):
    mkdir(dstDir)
```

```python
for dir in listdir(imgsDir):
    print(dir)
    cont = 0
    if not isdir(dstDir+dir):
        mkdir(dstDir+dir)
        
    for img in listdir(imgsDir+dir):
        readImg = cv2.imread(imgsDir+dir+'/'+img, 1)
        resized = cv2.resize(readImg, (width, height))
        #cv2.imshow('ciao', resized)
        cv2.imwrite(dstDir+dir+'/'+img, resized)
        cont += 1
        if cont % 10 == 0:
            print(cont)
```
